# K-Means Clustering

## Importing Libraries

## Plotting Make vs Displacement

# Example of K=2 Clusters

```python
Km = KMeans(n_clusters=2)
Km
```

```python
y_predicted=Km.fit_predict(dfle[['Make','Displacement']])
y_predicted
```

```python
dfle['clusters']=y_predicted
```

```python
df1=dfle[dfle['clusters']==0]
df2=dfle[dfle['clusters']==1]
plt.scatter(df1['Make'],df1['Displacement'],color='b')
plt.scatter(df2['Make'],df2['Displacement'],color='r')
plt.title('K-Clusters-2')
plt.show()
```

```python
scaler=MinMaxScaler()
scaler.fit(dfle[['Displacement']])
dfle['Displacement']=scaler.transform(dfle[['Displacement']])
scaler.fit(dfle[['Make']])
dfle['Make']=scaler.transform(dfle[['Make']])
```

```python
Km=KMeans(n_clusters=2)
y_predicted=Km.fit_predict(dfle[['Make','Displacement']])
dfle['newclusters']=y_predicted
ndf1=dfle[dfle['newclusters']==0]
ndf2=dfle[dfle['newclusters']==1]
plt.scatter(ndf1['Make'],ndf1['Displacement'],color='b')
plt.scatter(ndf2['Make'],ndf2['Displacement'],color='r')
plt.title('K-Clusters-2')
plt.show()
```

```python
centers=Km.cluster_centers_
centers
```

```python
Km=KMeans(n_clusters=2)
y_predicted=Km.fit_predict(dfle[['Make','Displacement']])
dfle['newclusters']=y_predicted
ndf1=dfle[dfle['newclusters']==0]
ndf2=dfle[dfle['newclusters']==1]
plt.scatter(ndf1['Make'],ndf1['Displacement'],color='b')
plt.scatter(ndf2['Make'],ndf2['Displacement'],color='r')
plt.scatter(centers[:,0],centers[:,1],color='purple',marker='o',label='centroid')
plt.title('K-Clusters-2')
plt.show()
```

## Finding Best value of K using algo

```python
k_rng=range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(dfle[['Make','Displacement']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse,color='gold')
plt.show
```

## Clustering at best value of K 

```python
##elblow at 4 
Km=KMeans(n_clusters=4)
y_predicted=Km.fit_predict(dfle[['Make','Displacement']])
dfle['newclusters']=y_predicted
nndf1=dfle[dfle['newclusters']==0]
nndf2=dfle[dfle['newclusters']==1]
nndf3=dfle[dfle['newclusters']==2]
nndf4=dfle[dfle['newclusters']==3]
centers=Km.cluster_centers_
plt.scatter(nndf1['Make'],nndf1['Displacement'],color='b')
plt.scatter(nndf2['Make'],nndf2['Displacement'],color='r')
plt.scatter(nndf3['Make'],nndf3['Displacement'],color='g')
plt.scatter(nndf4['Make'],nndf4['Displacement'],color='y')
plt.scatter(centers[:,0],centers[:,1],color='purple',marker='*',label='centroid')
plt.title('K-Clusters-4')
plt.show()
```
